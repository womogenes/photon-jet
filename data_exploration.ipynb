````python
def get_numpy(dataset):
    data_dir = "/usatlas/atlas01/atlasdisk/users/atlas_wifeng/photon-jet/data"
    f = h5py.File(f"{data_dir}/raw_files/h5/{dataset}")
    return dict(f)
````

## Get some jet images for $\pi_0$

````python
for key in no_mass:
    print(key.ljust(20), no_mass[key].shape)
````

````python
# Investigate total energy across labels vs energy
M = 100000

tot_energy = np.zeros(M)
for layer in ["layer_0", "layer_1", "layer_2", "layer_3"]:
    reshaped = np.reshape(no_mass[layer], (M, -1))
    tot_energy += np.sum(reshaped, axis=1)
    
tot_energy += np.sum(np.sum(no_mass["overflow"], axis=1))
````

````python
plt.scatter(no_mass["energy"], tot_energy);
plt.xlabel("Sum of energies in calo cells")
plt.ylabel("\"energy\" feature (column?)");
````

## Get jet images for separate jet types

````python
# Combine everything in layer_1
def graph_total(dataset, layer, ax):
    sum_layer = np.sum(dataset[layer], axis=0)
    
    ax.imshow(
        np.log(sum_layer.T),
        interpolation='none',
        cmap="viridis",
        extent=[-0.2, 0.2, -0.02, 0.02],
        aspect=10
    )
````

````python
def graph_all_layers(dataset_name):
    dataset = get_numpy(dataset_name)
    
    fig, axs = plt.subplots(1, 4, figsize=(20, 4))
    for layer, ax in zip(["layer_0", "layer_1", "layer_2", "layer_3"], axs):
        graph_total(dataset, layer, ax)
        ax.set_title(layer)
        
    fig.suptitle(dataset_name)
````

````python
graph_all_layers("pi0_40-250GeV_100k.h5")
````

````python
graph_all_layers("gamma_40-250GeV_100k.h5")
````

````python
graph_all_layers("scalar1_40-250GeV_100k.h5")
````
